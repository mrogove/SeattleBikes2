```python
import pandas,sqlite3
```

```python
df1 = pandas.read_csv('counterData/Elliott_Bay_Trail_in_Myrtle_Edwards_Park.csv',
                        usecols = ["Date","Bike North","Bike South"],
                        header = 0,
                     )
```

```python
df1.rename(columns=lambda x: x.lstrip(), inplace=True) #strip column names
```

```python
df1.rename(columns={'Bike North':'NB','Bike South':'SB'}, inplace=True) #rename for total count
```

```python
df1["TotalCount"]=df1["NB"]+df1["SB"]
```

```python
df1["Name"] = "ElliotBayTrailMyrtlePark"
```

```python
df1["Latitude"]  = 47.6194807
df1["Longitude"] = -122.361993
```

```python
df1 = df1[['Name','Date', 'Latitude', 'Longitude', 'NB', 'SB', 'TotalCount']]
```

```python
db = sqlite3.connect('cleanedCounterData/bikes.db')
df1.to_sql('BIKECOUNT', db, if_exists='append', index=False)
```

```python
df1.to_csv('cleanedCounterData/ElliotBayTrailMyrtleEdwards.csv', float_format="%.0f")
```

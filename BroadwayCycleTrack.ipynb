```python
import pandas,sqlite3
```

```python
df1 = pandas.read_csv('counterData/Broadway_Cycle_Track_North_Of_E_Union_St.csv',
                        usecols = ["Date","Broadway Cycle Track North Of E Union St Total","NB","SB"],
                        header = 0,
                     )
```

```python
df1.rename(columns=lambda x: x.lstrip(), inplace=True) #strip column names
```

```python
df1.rename(columns={'Broadway Cycle Track North Of E Union St Total': 'TotalCount', 'North':'NB','South':'SB'}, inplace=True) #rename for total count
```

```python
df1["Name"] = "BroadwayCycleTrack"
```

```python
df1["Latitude"]  =  47.613042
df1["Longitude"] = -122.320776
```

```python
df1 = df1[['Name','Date', 'Latitude', 'Longitude', 'NB', 'SB', 'TotalCount']]
```

```python
db = sqlite3.connect('cleanedCounterData/bikes.db')
df1.to_sql('BIKECOUNT', db, if_exists='append', index=False)
```

```python
df1.to_csv('cleanedCounterData/BroadwayCycleTrack.csv')
```

```python
import pandas,sqlite3
```

```python
df1 = pandas.read_csv('counterData/39th_Ave_NE_Greenway_at_NE_62nd_St.csv',
                        usecols = ["Date","39th Ave NE Greenway at NE 62nd St Total","North","South"],
                        header = 0,
                     )
```

```python
df1.rename(columns=lambda x: x.lstrip(), inplace=True) #strip column names
```

```python
df1.rename(columns={'39th Ave NE Greenway at NE 62nd St Total': 'TotalCount','North':'NB','South':'SB'}, inplace=True) #rename for total count
```

```python
df1["Name"] = "39thAveGreenway"
```

```python
df1["Latitude"]  = 47.673968
df1["Longitude"] = -122.285785
```

```python
df1 = df1[['Name','Date', 'Latitude', 'Longitude', 'NB', 'SB', 'TotalCount']] #reorder cols
```

```python
db = sqlite3.connect('cleanedCounterData/bikes.db')
df1.to_sql('BIKECOUNT', db, if_exists='append', index=False)
```

```python
df1.to_csv('cleanedCounterData/39thAveGreenway.csv')
```

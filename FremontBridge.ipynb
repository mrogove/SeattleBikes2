```python
import pandas,sqlite3
```

```python
df1 = pandas.read_csv('counterData/Fremont_Bridge_Hourly_Bicycle_Counts_by_Month_October_2012_to_present.csv',
                        usecols = ["Date","Fremont Bridge West Sidewalk","Fremont Bridge East Sidewalk"],
                        header = 0,
                     )
```

```python
df1.rename(columns=lambda x: x.lstrip(), inplace=True) #strip column names
```

```python
df1.rename(columns={"Fremont Bridge West Sidewalk":"SB","Fremont Bridge East Sidewalk":"NB"}, inplace=True) #rename for total count
```

```python
df1["Name"] = "FremontBridge"
```

```python
df1["Latitude"]  = 47.6475229
df1["Longitude"] = -122.3508511
```

```python
df1["TotalCount"] = df1["NB"]+df1["SB"]
```

```python
df1 = df1[['Name','Date', 'Latitude', 'Longitude', 'NB', 'SB', 'TotalCount']]
```

```python
db = sqlite3.connect('cleanedCounterData/bikes.db')
df1.to_sql('BIKECOUNT', db, if_exists='append', index=False)
```

```python
df1.to_csv('cleanedCounterData/FremontBridge.csv', float_format="%.0f")
```

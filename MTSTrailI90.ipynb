```python
import pandas,sqlite3
```

```python
df1 = pandas.read_csv('counterData/MTS_Trail_west_of_I-90_Bridge.csv',
                        usecols = ["Date","Bike East","Bike West"],
                        header = 0,
                     )
```

```python
df1.rename(columns=lambda x: x.lstrip(), inplace=True) #strip column names
```

```python
df1.rename(columns={'Bike East':'NB','Bike West':'SB'}, inplace=True) #rename for total count
```

```python
df1["TotalCount"]=df1["NB"]+df1["SB"]
```

```python
df1["Name"] = "MTSTrailI90"
```

```python
df1["Latitude"]  = 47.590444
df1["Longitude"] = -122.288869
```

```python
df1 = df1[['Name','Date', 'Latitude', 'Longitude', 'NB', 'SB', 'TotalCount']]
```

```python
db = sqlite3.connect('cleanedCounterData/bikes.db')
df1.to_sql('BIKECOUNT', db, if_exists='append', index=False)
```

```python
df1.to_csv('cleanedCounterData/MTSTrailI90.csv')
```
